```python
# Read the CSV file britishSeatBeltStudy.csv
import pandas as pd
df = pd.read_csv("/content/gdrive/MyDrive/mystuffpgss2022/Lecture2/britishSeatBeltStudy.csv")
df['law'] = df['law'].astype('category')
df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m/%d')
df
```

# Find statistics from data

```python
df.boxplot('DriversKilled', 'law')
```

```python
df.plot.scatter(x = 'Date', y = 'DriversKilled', c = 'law', colormap = 'viridis')
```

# Add column for fraction of drivers who are killed

```python
df['DeathRatio'] = df['DriversKilled'] / df['drivers']
```

```python
df.plot.scatter(x = 'Date', y = 'DeathRatio', c = 'law', colormap = 'viridis')
```

```python
df.boxplot('DeathRatio', 'law')
```

# Examine specific groups of deaths

```python
df['VanRatio'] = df['VanKilled'] / df['drivers']
df['FrontRatio'] = df['front'] / df['drivers']
df['RearRatio'] = df['rear'] / df['drivers']
```

```python
df.plot.scatter(x = 'Date', y = 'VanRatio', c = 'law', colormap = 'viridis')
```

```python
df.plot.scatter(x = 'Date', y = 'FrontRatio', c = 'law', colormap = 'viridis')
```

```python
df.plot.scatter(x = 'Date', y = 'RearRatio', c = 'law', colormap = 'viridis')
```
